#*Scraping*

***Extracción del código***

```python
# Link de Mercado Libre
url_page = 'https://ofertas.mercadolibre.com.mx/flagship-computo-gaming#deal_print_id=da101360-6c55-11ed-91fb-2de9b38d5c14&c_id=special-normal&c_element_order=2&c_campaign=GAMING&c_uid=da101360-6c55-11ed-91fb-2de9b38d5c14'
page = requests.get(url_page).text 
soup = BeautifulSoup(page, "lxml")
```

***Extracción de los Link de articulos***

```python
# Extraemos los links de las publicacones
a = soup.find_all('a')
l = []
for i in a:
    if 'https://articulo.mercadolibre.com.mx' in str(i):
        l.append(str(i))
```
